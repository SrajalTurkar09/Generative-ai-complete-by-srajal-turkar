```python
# Free CLIP Model Example (Hugging Face version)
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import requests

# Load a pre-trained CLIP model and processor (open-source)
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Example image and text data
image_url = "https://huggingface.co/datasets/nateraw/food/raw/images/pizza.jpg"
image = Image.open(requests.get(image_url, stream=True).raw)

# Input text descriptions to match with the image
texts = ["A photo of a pizza", "A photo of a cat"]

# Preprocess the image and text for CLIP model
inputs = processor(text=texts, images=image, return_tensors="pt", padding=True)

# Get the similarity scores between the image and text
outputs = model(**inputs)
logits_per_image = outputs.logits_per_image  # Image-to-text matching scores
probs = logits_per_image.softmax(dim=1)  # Probabilities of matching

# Print the probabilities of each text matching the image
print("Probabilities of matching text to image:", probs)
```

```python
# Free CLIP Model Example (Hugging Face version)
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import requests
import io

# Load a pre-trained CLIP model and processor (open-source)
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Example image and text data
# This URL points to a public image that should be accessible without authentication
image_url = "https://content.jdmagicbox.com/comp/def_content/pizza_outlets/default-pizza-outlets-13.jpg"

# Download the image content and wrap it in a BytesIO object
response = requests.get(image_url, stream=True)
response.raise_for_status()  # This will raise an exception if the request was unsuccessful
image_content = response.content
image = Image.open(io.BytesIO(image_content))

# Input text descriptions to match with the image
texts = ["A photo of a pizza", "A photo of a cat"]

# Preprocess the image and text for CLIP model
inputs = processor(text=texts, images=image, return_tensors="pt", padding=True)

# Get the similarity scores between the image and text
outputs = model(**inputs)
logits_per_image = outputs.logits_per_image  # Image-to-text matching scores
probs = logits_per_image.softmax(dim=1)  # Probabilities of matching

# Print the probabilities of each text matching the image
print("Probabilities of matching text to image:", probs)
```
